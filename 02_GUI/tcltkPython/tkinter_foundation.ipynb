```python

# Addtion of Two Number
# GUI Application
import tkinter as tk
root = tk.Tk()
root.title("Calc")
root.geometry("300x400")

def add_number():
    try:
        val = int(entry1.get()) + int(entry2.get())
        re.config(text=f"Result is: {val}")
    except ValueError:
        re.config(text=f"Invalid Values")


   

tk.Label(root, text="Enter the First Number").pack(pady=5)

entry1 = tk.Entry(root)
entry1.pack(pady=5)


tk.Label(root, text="Enter the Second Number").pack()

entry2 = tk.Entry(root)
entry2.pack()

bt = tk.Button(root, text="Add", command=add_number)
bt.pack()

re = tk.Label(root, text="Result is: ", font=('arial', 12))
re.pack()

root.mainloop()
```
